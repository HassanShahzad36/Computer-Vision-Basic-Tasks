```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
def negativeTransformation(img):
    originalImg = cv.imread(img,cv.IMREAD_GRAYSCALE)
    L = pow(2 , (originalImg.itemsize*8))
    negativeImg = L-1-originalImg

    fig,axes = plt.subplots(1,2,figsize=(10,5))
    axes[0].imshow(originalImg, cmap='gray')
    axes[0].set_title("Original Image")
    axes[1].imshow(negativeImg,cmap='gray')
    axes[1].set_title("Negetive Transformed Image")
    
    plt.show()
negativeTransformation("Mona Lisa.jpg")
```

```python
def logTransforamtion(img):
    base = int(input("Enter Base of logarithm"))
    originalImg = cv.imread(img,cv.IMREAD_GRAYSCALE)
    logTransformedImg = np.log(1+originalImg)/np.log(base) #suppose c = 1 
    logTransformedImg = logTransformedImg - np.min(logTransformedImg)  # Shift to zero
    logTransformedImg = logTransformedImg / np.max(logTransformedImg)  # Normalize to [0, 1]
    logTransformedImg = np.round(logTransformedImg * 255)  # Scale to [0, 255]
    logTransformedImg = logTransformedImg.astype(np.uint8)  # Convert to uint8
  
    
    
    fig,axes = plt.subplots(1,2,figsize=(10,5))
    axes[0].imshow(originalImg,cmap='gray')
    axes[0].set_title("Original Image")
    axes[1].imshow(logTransformedImg,cmap='gray')
    axes[1].set_title("Log Transformed Image")
    plt.show() 

logTransforamtion("Mona Lisa.jpg")
```

```python
def powerTransformation(img,c=1):
    gamma = float(input("Enter gamma value"))
    originalImg = cv.imread(img,cv.IMREAD_GRAYSCALE)

    gammaCorrectedImg =c*(255*(originalImg / 255) ** gamma)
    gammaCorrectedImg = gammaCorrectedImg.astype(np.uint8)

    fig,axes = plt.subplots(1,2,figsize=(10,5))
    axes[0].imshow(originalImg,cmap='gray')
    axes[0].set_title("Original Image")
    axes[1].imshow(gammaCorrectedImg,cmap='gray')
    axes[1].set_title("Gamma Corrected Image")
    plt.show()

powerTransformation("Mona Lisa.jpg")

    
```

```python
def contrastStretching(img):
    min = int(input("Minimum Gray Scale Level"))
    max = int(input("Maximum Gray Scale Level"))

    originalImg = cv.imread(img,cv.IMREAD_GRAYSCALE)
    contrastStretchedImg = cv.normalize(originalImg,None,alpha=min,beta=max,norm_type=cv.NORM_MINMAX)

    fig,axes = plt.subplots(1,2,figsize=(10,5))
    axes[0].imshow(originalImg,cmap='gray')
    axes[0].set_title("Original Image")
    axes[1].imshow(contrastStretchedImg,cmap='gray')
    axes[1].set_title("Contrast Stretch Image")
    plt.show() 

contrastStretching("Mona Lisa.jpg")
    
     
```

```python
def grayLevelSlicing(img):
    min = int(input("Enter minimum gray scale"))
    max = int(input("Enter maximum gray scale"))
    originalImg = cv.imread(img, cv.IMREAD_GRAYSCALE)
    grayScaleSlicedImg = np.copy(originalImg)

    for i in range(originalImg.shape[0]):
        for j in range(originalImg.shape[1]):
            if min<= originalImg[i, j] and max>= originalImg[i,j]:
                grayScaleSlicedImg[i, j] = 255
            else:
                grayScaleSlicedImg[i,j] = originalImg[i,j]

    fig,axes = plt.subplots(1,2,figsize=(10,5))

    axes[0].imshow(originalImg,cmap='gray')
    axes[0].set_title("Original Image")
    axes[1].imshow(grayScaleSlicedImg,cmap='gray')
    axes[1].set_title("Gray Scale Sliced Image")
    plt.show()

grayLevelSlicing("Mona Lisa.jpg")
    


```

```python
def bitSlicing(img):
    originalImg = cv.imread(img, cv.IMREAD_GRAYSCALE)
    bits = int(input("How many bits to be extraced"))

    if(bits>=0 and bits<=7):
        bitPlaneImg = (originalImg>> bits) & 1
        bitPlaneImg =255* bitPlaneImg

        fig,axes = plt.subplots(1,2,figsize=(10,5))

        axes[0].imshow(originalImg,cmap='gray')
        axes[0].set_title("Original Image")
        axes[1].imshow(bitPlaneImg,cmap='gray')
        axes[1].set_title("Bit Plane Sliced Image")
        plt.show()
    else:
        return 
bitSlicing("Mona Lisa.jpg")
    

```
