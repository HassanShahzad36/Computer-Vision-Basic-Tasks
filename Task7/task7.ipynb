```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
from scipy.signal import convolve2d
```

```python
def edge_detection(image,method):
    if(method=="sobel"):
        sobelX = np.array([[-1,0,1],
                          [-2,0,2],
                          [-1,0,1]])
        
        sobelY = np.array([[-1,-2,-1],
                           [0,0,0],
                           [1,2,1]])
        
        gradientX = convolve2d(image,sobelX,mode='same')
        gradientY = convolve2d(image,sobelY,mode='same')
        gradient = np.sqrt(gradientX**2+gradientY**2)
        print(gradient)

        plt.imshow(gradient,cmap="gray")
        plt.axis("off")
        plt.title("Sobel Edge Detector")

    elif(method=='prewitt'):
         prewittX = np.array([[-1,0,1],
                          [-1,0,1],
                          [-1,0,1]])
        
         prewittY = np.array([[-1,-1,-1],
                           [0,0,0],
                           [1,1,1]])
        
         gradientX = convolve2d(image,prewittX,mode='same')
         gradientY = convolve2d(image,prewittY,mode='same')
         gradient = np.sqrt(gradientX**2+gradientY**2)

         plt.imshow(gradient,cmap="gray")
         plt.axis("off")
         plt.title("Prewitt Edge Detector")

    elif(method=='roberts'):
         robertsX = np.array([[1,0],
                            [0,-1]])
        
         roberstY = np.array([[0,1],
                              [-1,0]])
        
         gradientX = convolve2d(image,robertsX,mode='same')
         gradientY = convolve2d(image,roberstY,mode='same')
         gradient = np.sqrt(gradientX**2+gradientY**2)

         plt.imshow(gradient,cmap="gray")
         plt.axis("off")
         plt.title("Roberts Edge Detector")


        
```

```python
img = cv.imread("Chiron.jpeg")
img = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
print(edge_detection(img,"sobel"))
# print(edge_detection(img,"prewitt"))
# print(edge_detection(img,"roberts"))
```
